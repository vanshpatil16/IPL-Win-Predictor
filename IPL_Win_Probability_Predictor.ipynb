```python
import pandas as pd
import numpy as np
```

```python
match=pd.read_csv('matches.csv')
delivery=pd.read_csv('deliveries.csv')
```

```python
match.head()
```

```python
delivery.head()
```

```python
total_score_df=delivery.groupby(['match_id','inning']).sum()['total_runs'].reset_index()
```

```python
total_score_df
```

```python
total_score_df=total_score_df[total_score_df['inning']==1]
```

```python
total_score_df.head()
```

```python
match_df=match.merge(total_score_df[['match_id','total_runs']],left_on='id',right_on='match_id')
```

```python
match_df.head(10)
```

```python
match_df['team1'].unique()
```

```python
teams=['Royal Challengers Bangalore', 'Kings XI Punjab',
        'Mumbai Indians', 'Kolkata Knight Riders',
       'Rajasthan Royals',  'Chennai Super Kings',
       'Pune Warriors', 'Sunrisers Hyderabad',
        'Delhi Capitals']
```

```python
match_df['team1']=match_df['team1'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team2']=match_df['team2'].str.replace('Delhi Daredevils', 'Delhi Capitals')
match_df['team1']=match_df['team1'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
match_df['team2']=match_df['team2'].str.replace('Deccan Chargers', 'Sunrisers Hyderabad')
```

```python
match_df=match_df[match_df['team1'].isin(teams)]
match_df=match_df[match_df['team2'].isin(teams)]
```

```python
 match_df.info()
```

```python
match_df=match_df[['match_id','winner','total_runs']]
```

```python
delivery_df=match_df.merge(delivery,on='match_id')
delivery_df
```

```python
delivery_df=delivery_df[delivery_df['inning']==1]
```

```python
delivery_df.head(10)
```

```python
delivery_df.info()
```

```python
print(delivery_df['total_runs_y'].dtype)
    
```

```python
delivery_df['current_score']=delivery_df.groupby('match_id')['total_runs_y'].cumsum()
```

```python
delivery_df.head(20)
```

```python
delivery_df['runs_left']=delivery_df['total_runs_x']-delivery_df['current_score']
```

```python
delivery_df
```

```python
delivery_df['over'].unique()
```

```python
delivery_df['balls_left']=120-(delivery_df['over']*6 + delivery_df['ball'])
```

```python
delivery_df.head(20)
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].fillna("0")
```

```python
delivery_df['player_dismissed']
```

```python
delivery_df['player_dismissed']=delivery_df['player_dismissed'].apply(lambda x:x if x=="0" else "1")
```

```python

delivery_df['player_dismissed']=delivery_df['player_dismissed'].astype('int')
wickets=delivery_df.groupby('match_id')['player_dismissed'].cumsum().values
delivery_df['wickets']=10-wickets
delivery_df.head(20)
                                                                      
```

```python
#crr(current run rate)=runs/overs
delivery_df['crr']=(delivery_df['current_score']*6)/(120-delivery_df['balls_left'])
delivery_df.head()
```

```python
#rrr(required run rate)=runs left/overs_left
delivery_df['rrr']=(delivery_df['runs_left']*6)/(delivery_df['balls_left'])
```

```python
delivery_df.head()
```

```python
def result(row):
    return 1 if row['batting_team']==row['winner'] else 0
```

```python
delivery_df['result']=delivery_df.apply(result,axis=1)
delivery_df['result']
```

```python
final_df=delivery_df[['batting_team','bowling_team','runs_left','balls_left','wickets','total_runs_x','crr','rrr','result']]
```

```python
final_df
```

```python
final_df.sample()
```

```python
final_df.isnull().sum()
```

```python
final_df=final_df[final_df['balls_left']!=0]
```

```python
final_df.dropna(inplace=True)
```

```python
X=final_df.drop('result',axis=1)
X
```

```python
y=final_df['result']
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([
    ('trf',OneHotEncoder(sparse_output=False,drop='first'),['batting_team','bowling_team'])
]
                      ,remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe=Pipeline(steps=[
    ('step1',trf),
    ##('step2',LogisticRegression(solver='liblinear')),
    ('step2',RandomForestClassifier())
])
```

```python
pipe.fit(X_train,y_train)
```

```python
X_train.describe()
```

```python
y_pred=pipe.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
